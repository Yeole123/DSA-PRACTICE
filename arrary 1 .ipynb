```python
# Function to transpose a matrix
def transpose_matrix(matrix, rows, cols):
    transpose = [[0 for _ in range(rows)] for _ in range(cols)]

    for i in range(rows):
        for j in range(cols):
            transpose[j][i] = matrix[i][j] 

    return transpose


matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

rows = len(matrix)      
cols = len(matrix[0])   


result = transpose_matrix(matrix, rows, cols)


print("Original Matrix:")
for row in matrix:
    print(row)


print("\nTransposed Matrix:")
for row in result:
    print(row)
```

```python
def binarysearch(array, i, j, x):
    while i <= j:
        mid = i + (j - i) // 2
        if array[mid] < x:
            return binarysearch(array, mid + 1, j, x)  # Return the result of recursive call
        elif array[mid] > x:
            return binarysearch(array, i, mid - 1, x)  # Return the result of recursive call
        else:
            return mid
    return -1

# Driver code 
array = [1, 2, 3, 4]
i = 0
j = len(array) - 1
x = 4
print(binarysearch(array, i, j, x))  # Output: 3
```

```python
def getMax(operations):
    stack = [0] * len(operations)
    max_stack = [0] * len(operations)
    result = [0] * len(operations)
    stack_top = -1
    max_top = -1
    result_size = 0

    for op in operations:
        parts = op.split()
        command = int(parts[0])

        if command == 1:
            val = int(parts[1])
            stack_top += 1
            stack[stack_top] = val

            if max_top == -1 or val >= max_stack[max_top]:  
                max_top += 1
                max_stack[max_top] = val

        elif command == 2:
            if stack_top >= 0:
                popped = stack[stack_top]
                stack_top -= 1

                if max_top >= 0 and popped == max_stack[max_top]:  
                    max_top -= 1

        elif command == 3:
            if max_top >= 0:
                result[result_size] = max_stack[max_top]
                result_size += 1

    final_result = [result[i] for i in range(result_size)]
    return final_result
```

```python
# binary searchin 2d matrix 
def searchsort(matrix,target):
    m=len(matrix)
    if m==0:
        return False
    n=len(matrix[0])
    left,right=0,m*n-1
    while left<=right:
        mid=left+(right-left)//2
        mid_element=matrix[mid//n][mid%n] 
        if target==mid_element:
            return True
        elif target<mid_element:
            right=mid-1
        else:
            left=mid+1
    return False 


matrix=[[1,3,5,7],[10,11,16,20],[23,30,34,60]]
target=6
result=searchsort(matrix,target)
print(result)
```

```python
# ternary search: divides the whole search part in two search spaces 
def ternarysearch(left,right,x,arr):
    while left<=right:
        mid1=left+(right-left)//3
        mid2=right-(right-left)//3
        if arr[mid1]==x:
            return mid1
        elif arr[mid2]==x:
            return mid2
        elif x>arr[mid2]:
            return ternarysearch(mid2+1,right,x,arr)
        elif x<arr[mid1]:
            return ternarysearch(left,mid1-1,x,arr)
        else:
            return ternarysearch(mid1+1,mid2-1,x,arr)
    return -1
arr=[1,2,3,4,5,6,7,8]
left=0
right=len(arr)-1
x=5
result=ternarysearch(left,right,x,arr)
print(f"searching element is present at index: {result}")
```
